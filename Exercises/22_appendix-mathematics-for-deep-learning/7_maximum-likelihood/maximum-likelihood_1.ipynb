#### 1. Suppose that you know that a non-negative random variable has density $\alpha e^{-\alpha x}$ for some value $\alpha > 0$. You obtain a single observation from the random variable which is the number $3$. What is the maximum likelihood estimate for $\alpha$?

Note that
$$
\begin{aligned}
E\left[\alpha e^{-\alpha x}\right] &= \alpha E\left[e^{-\alpha x}\right] \\
&= \alpha \int_0^\infty x e^{- \alpha x} dx \\
&= \alpha \left.\frac{e^{-\alpha x}(\alpha x + 1)}{\alpha^2}\right|_{x = 0}^{x = \infty} \\
&= \alpha \frac{1}{\alpha^2} = \frac{1}{\alpha}.
\end{aligned}
$$
As such the maxium likelihood estimator is $\alpha = \frac{1}{3}$.

```python
import matplotlib.pyplot as plt
import numpy as np

def func(x, alpha):
    return alpha * np.exp(-alpha * x)

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Plot 1
for alpha in [0.1, 0.2, 1/3, 0.5, 1]:
    x = np.linspace(2, 4, 200)
    y = func(x, alpha)
    ax1.plot(x, y, label=f"$\\alpha$={alpha:.3f}")

ax1.axvline(x=3, color='black', linestyle='--')
ax1.set_title("Plot of $f_{\\alpha}(x) = \\alpha e^{-\\alpha x}$")
ax1.set_xlabel("x")
ax1.set_ylabel("f(x)")
ax1.legend()

# Plot 2
x_alpha = np.linspace(0, 1, 200)
y = func(3, x_alpha)
ax2.plot(x_alpha, y, label="$f_{\\alpha}(3)$")
ax2.axvline(x=1/3, color='r', linestyle='--')
ax2.set_title("Plot of $f_{\\alpha}(3)$")
ax2.set_xlabel("$\\alpha$")
ax2.set_ylabel("$f_{\\alpha}(3)$")
ax2.legend()

# Set the same y-limits for both subplots
y_min = min(min(y) for y in [func(x, alpha) for alpha in [0.1, 0.2, 1/3, 0.5, 1]])
y_max = max(max(y) for y in [func(x, alpha) for alpha in [0.1, 0.2, 1/3, 0.5, 1]])
ax1.set_ylim(y_min, y_max)
ax2.set_ylim(y_min, y_max)

plt.tight_layout()
plt.savefig("maximum-likelihood_1_1.png")
plt.show()
```
